```python
import pandas as pd
import numpy as np
import ta
import logging
import joblib
import json
from datetime import datetime, timedelta
from pathlib import Path
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import RobustScaler, StandardScaler
from catboost import CatBoostClassifier
from typing import Dict, Tuple
from sklearn.model_selection import TimeSeriesSplit, learning_curve
from sklearn.metrics import f1_score
from sklearn.compose import ColumnTransformer
from sklearn.utils.class_weight import compute_class_weight
from sklearn.feature_selection import RFECV
import matplotlib.pyplot as plt
import optuna
import sys
from dateutil.relativedelta import relativedelta
from sklearn.base import BaseEstimator, TransformerMixin

class TypeCaster(BaseEstimator, TransformerMixin):
    """Cast specified columns back to int32 after the ColumnTransformer."""
    def __init__(self, cat_idx):
        self.cat_idx = cat_idx
    def fit(self, X, y=None):
        return self
    def transform(self, X):               
        X = X.copy()
        # 2) force object dtype so CatBoost will accept cat_features
        X = X.astype(object)
        # 3) overwrite only your categorical-feature columns with int codes
        for idx in self.cat_idx:
            # take the entire column, astype to int, then re-assign
            col = X[:, idx].astype(np.int32)
            X[:, idx] = col
        return X

# ==================== CONFIG ====================

class StrategyConfig:
    RISK_PER_TRADE: float = 0.005
    PARTIAL_CLOSE_PCT: float = 0.7
    MAX_DD: float = 0.04  # Daily drawdown threshold
    TARGET_THRESHOLD: float = 0.005 # Threshold for 3-class target

    # Unified instruments for 1H timeframe
    INSTRUMENTS: Dict[str, Dict] = {
       
         'EURUSD.4H': {'file': 'EURUSD_H4_200901020400_202506301200_cleaned.csv', 'pip': 0.0001, 'pip_value': 10},
      
    }

    # Model and scaler paths for unified strategy
    MODEL_PATH = 'eurusd4h_final_wfv54.pkl'
    SCALER_PATH = 'eurusd_4h_v52_scaler.pkl'

    # Training range
    TRAIN_START: datetime = datetime(2018, 1, 1)
    TRAIN_END: datetime = datetime(2024, 12, 31)

# -------------------- Helper for Risk Factor --------------------  
def get_risk_factor(symbol: str) -> float:  
    """  
    Returns 1% risk if symbol indicates a daily timeframe (contains "1D"),  
    else returns 0.5% risk.  
    """  
    return 0.01 if "1D" in symbol else 0.002  

# ==================== DATA ENGINE ====================
class LeanDataEngine:  
    def __init__(self):  
        self.instruments = self._load_clean_data()  

    def _load_clean_data(self) -> Dict[str, pd.DataFrame]:  
        instruments = {}  
        dtype_spec = {'Open': 'float32', 'High': 'float32', 'Low': 'float32',  
                      'Close': 'float32', 'Volume': 'float32'}  
        for sym, config in StrategyConfig.INSTRUMENTS.items():  
            df = pd.read_csv(config['file'], dtype=dtype_spec)  
            if 'Date' in df.columns:  
                df['Date'] = pd.to_datetime(df['Date'], errors='coerce')  
                df = df.dropna(subset=['Date'])  
                df.set_index('Date', inplace=True)  
            elif 'Datetime' in df.columns:  
                df['Datetime'] = pd.to_datetime(df['Datetime'], errors='coerce')  
                df = df.dropna(subset=['Datetime'])  
                df.set_index('Datetime', inplace=True)  
            df.sort_index(inplace=True)  
            df.index = pd.to_datetime(df.index, errors='coerce')  
            df = self._calculate_core_features(df)  
            instruments[sym] = df  
        return instruments  

    def _calculate_core_features(self, df: pd.DataFrame) -> pd.DataFrame:
        df['EMA20'] = ta.trend.ema_indicator(df['Close'], 20)
        df['EMA50'] = ta.trend.ema_indicator(df['Close'], 50)  # New: 50-period EMA
        df['EMA200'] = ta.trend.ema_indicator(df['Close'], 200)
        # New features based on EMA50
        df['ema50_diff'] = df['EMA20'] - df['EMA50']
        df['EMA50_slope'] = df['EMA50'].diff(5) / df['EMA50'].shift(5)
        
        df['BB_upper'] = ta.volatility.bollinger_hband(df['Close'], 20, 2)
        df['BB_lower'] = ta.volatility.bollinger_lband(df['Close'], 20, 2)
        # Add BB width calculation
        df['BB_width'] = (df['BB_upper'] - df['BB_lower']) / df['Close']
        # Add Lower Band penetration like the upper band one
        df['bb_lower_penetration'] = (df['Close'] - df['BB_lower']) / df['BB_lower']
        # right after you compute BB_upper / BB_lower / BB_width:
        df['bb_penetration'] = (df['Close'] - df['BB_upper']) / df['BB_upper']

        df['ATR'] = ta.volatility.average_true_range(df['High'], df['Low'], df['Close'], 14)
        df['ATR_MA'] = df['ATR'].rolling(20).mean()
        # Removed ATR_median100 from here
        # df['ATR_median100'] = df['ATR'].rolling(100).median()
        

        df['ADX'] = ta.trend.ADXIndicator(df['High'], df['Low'], df['Close'], window=14).adx()
        df['Trend_purity'] = df['Close'].rolling(20).corr(df['EMA20'].shift(1))
        df['EMA200_slope'] = df['EMA200'].diff(10) / df['EMA200'].shift(10)
        df['STD_20'] = df['Close'].rolling(20).std()
        df['Vol_ratio'] = df['ATR'] / df['STD_20']
        macd = ta.trend.MACD(df['Close'])
        df['MACD_hist'] = macd.macd_diff()

 
  

        df['vol_regime'] = np.where(
            df['ATR'] > df['ATR'].rolling(100).quantile(0.75),
            2,
            np.where(df['ATR'] > df['ATR'].rolling(100).quantile(0.4), 1, 0)
        )
        df['trend_volatility'] = df['ADX'] * df['ATR'] / df['Close']
        
        df['momentum_flag'] = np.sign(df['Close'].diff(5)) * ((abs(df['Close'].diff(5)) / df['ATR']) > 1).astype(int)
        # 1) RSI(14)
        df['RSI14'] = ta.momentum.rsi(df['Close'], window=14)


        

        # Volume features
        df['Volume_MA20'] = df['Volume'].rolling(20).mean()
        df['volume_ratio'] = df['Volume'] / df['Volume_MA20']
       


        return df.dropna()

# ==================== PERFORMANCE REPORT ====================
class PerformanceReport:
    def __init__(self, engine):
        self.trades = [t for t in engine.trades if t['status'] == 'closed']
        self.equity = engine.equity
        self.equity_timeseries = engine.equity_timeseries
        self.tp_after_partial_count = engine.tp_after_partial_count
        self.sl_after_partial_reverse_times = engine.sl_after_partial_reverse_times

    def _compute_sharpe(self):
        if self.equity_timeseries.empty:
            return 0
        daily_equity = self.equity_timeseries.resample('D').last().ffill()
        returns = daily_equity.pct_change().dropna()
        if returns.std() == 0:
            return 0
        sharpe = (returns.mean() / returns.std()) * np.sqrt(252)
        return np.clip(sharpe, -5, 5)

    def _compute_profit_factor(self):
        total_profit = 0
        total_loss = 0
        for trade in self.trades:
            if trade['direction'] == 'buy':
                pnl = (trade['exit_price'] - trade['entry_price']) * trade['size'] * trade['config']['pip_value']
            else:
                pnl = (trade['entry_price'] - trade['exit_price']) * trade['size'] * trade['config']['pip_value']
            if pnl > 0:
                total_profit += pnl
            elif pnl < 0:
                total_loss += abs(pnl)
        return total_profit / total_loss if total_loss > 0 else float('inf')

    def generate(self):
        # ----- true PnL-based win/loss counts -----
        def tpnl(t): return float(t.get('total_pnl', 0.0))
        pnl_losses = [t for t in self.trades if tpnl(t) < 0]
        pnl_wins   = [t for t in self.trades if tpnl(t) > 0]
        pnl_flat   = [t for t in self.trades if abs(tpnl(t)) < 1e-9]

        gross_profit = sum(tpnl(t) for t in pnl_wins)
        gross_loss   = -sum(tpnl(t) for t in pnl_losses)
        pf_pnl = (gross_profit / gross_loss) if gross_loss > 0 else float('inf')

        avg_win = (gross_profit / len(pnl_wins)) if pnl_wins else 0.0
        avg_loss = (-gross_loss / len(pnl_losses)) if pnl_losses else 0.0
        payoff = (avg_win / -avg_loss) if avg_loss != 0 else float('inf')

        print(f"\nPnL-based trades -> Wins: {len(pnl_wins)}  Losses: {len(pnl_losses)}  Flat: {len(pnl_flat)}")
        print(f"Avg Win: {avg_win:.2f}   Avg Loss: {avg_loss:.2f}   Payoff: {payoff:.2f}   PF (PnL-based): {pf_pnl:.2f}")

        total_return = (self.equity[-1] / 100000.0 - 1)
        peaks = np.maximum.accumulate(self.equity)
        dd = np.max((peaks - self.equity) / peaks)

        tp_full = sum(1 for t in self.trades if t['exit_reason'] == 'TP' and not t.get('partial_closed', False))
        tp_after_part = sum(1 for t in self.trades if t['exit_reason'] == 'TP' and t.get('partial_closed', False))
        sl_after_part = sum(1 for t in self.trades if t['exit_reason'] == 'SL' and t.get('partial_closed', False))
        sl_no_part = sum(1 for t in self.trades if t['exit_reason'] == 'SL' and not t.get('partial_closed', False))
        total = len(self.trades)

        # ===== HEADLINE STATS SHOULD BE PnL-BASED =====
        wins = len(pnl_wins)
        losses = len(pnl_losses)
        effective = wins + losses
        win_rate = wins / effective if effective > 0 else 0.0

        sharpe = self._compute_sharpe()
        pf = pf_pnl  # headline PF uses PnL-based calculation

        print(f"→ total trades: {total}")
        print(f"   TP w/o partial: {tp_full}")
        print(f"   TP after partial: {tp_after_part}")
        print(f"   SL after partial (still profit): {sl_after_part}")
        print(f"   SL w/o partial (true losses): {sl_no_part}\n")

        print(f"\n{' BACKTEST RESULTS ':=^40}")
        print(f"Initial Capital: $100,000.00")
        print(f"Final Equity: ${self.equity[-1]:,.2f}")
        print(f"Total Return: {total_return:.1%}")
        print(f"Max Drawdown: {dd:.1%}")
        print(f"Total Trades: {total}")
        print(f"Win Rate: {win_rate:.1%}   ({wins} wins, {losses} losses)")
        print(f"Sharpe Ratio: {sharpe:.2f}")
        print(f"Profit Factor: {pf:.2f}")
        print(f"TP hit after partial close: {self.tp_after_partial_count}")
        if self.sl_after_partial_reverse_times:
            avg_rev = np.mean(self.sl_after_partial_reverse_times)
            print(f"Average time from breakeven to SL (seconds): {avg_rev:.2f}")
        else:
            print("No trades hit SL after partial close.")
        self._report_trades_per_pair()
        self._plot_equity()

    def _report_trades_per_pair(self):
        trades_per_pair = {}
        for trade in self.trades:
            symbol = trade['symbol']
            trades_per_pair[symbol] = trades_per_pair.get(symbol, 0) + 1
        print("\nTrades per pair:")
        for symbol, count in trades_per_pair.items():
            print(f"{symbol}: {count}")

    def _plot_equity(self):
        plt.figure(figsize=(10, 6))
        plt.plot(self.equity)
        plt.title("Equity Curve")
        plt.xlabel("Time")
        plt.ylabel("Equity ($)")
        plt.grid(True)
        plt.show()
# ==================== BACKTEST DATA LOADER ====================
class BacktestDataLoader:
    def __init__(self):
        self.feature_order = None
        # Load pre-trained model and scaler
        self.bull_model = joblib.load(StrategyConfig.MODEL_PATH)
        self.feature_order = self.bull_model.named_steps['preprocessor'].feature_names_in_
        self.instruments = {}
        data_engine = LeanDataEngine()
        for sym, df in data_engine.instruments.items():
            self.instruments[sym] = {
                'data': df,
                'config': StrategyConfig.INSTRUMENTS[sym]
            }
        print("Backtest data loaded.")

# ==================== EXECUTION ENGINE ====================
class ExecutionEngine:
    def __init__(self):
        self.trades = []
        self.equity = [100000.0]
        self.open_trades = 0
        self.cooldowns = {}
        self.current_day = None
        self.daily_equity_start = None
        self.equity_timeseries = pd.Series([100000.0], index=[pd.Timestamp(datetime.now().date())])
        self.trade_counts = {}
        self.buy_count = 0
        self.sell_count = 0
        # New metrics:
        self.tp_after_partial_count = 0
        self.sl_after_partial_reverse_times = []  # list of time differences in seconds


    # ---- dynamic threshold helpers ----
    def get_threshold(self, symbol: str, direction: str, now):
        s = self.thresh[symbol][direction]
        if s['until'] is not None and now > s['until']:
            s['t'] = StrategyConfig.BASE_THRESH
            s['until'] = None
            s['streak'] = 0
        return float(s['t'])

    # ---- trading engine ----
    def update_equity_daily(self, current_dt: datetime):
        current_date = current_dt.date()
        last_date = self.equity_timeseries.index[-1].date()
        if current_date > last_date:
            last_equity = self.equity[-1]
            new_entry = pd.Series([last_equity], index=[pd.Timestamp(current_date)])
            self.equity_timeseries = pd.concat([self.equity_timeseries, new_entry])

    def execute_trade(self, signal: dict):
        # Enforce maximum of 6 concurrent open trades
        if self.open_trades >= 6:
            return

        if any(t['symbol'] == signal['symbol'] and t['status'] == 'open' for t in self.trades):
            return

        current_day = signal['datetime'].date()
        if self.current_day != current_day:
            self.current_day = current_day
            self.daily_equity_start = self.equity[-1]
        elif self.equity[-1] < self.daily_equity_start * (1 - StrategyConfig.MAX_DD):
            print(f"Daily drawdown reached on {signal['datetime'].date()}. No new trade for {signal['symbol']}.")
            return

        risk_value = get_risk_factor(signal['symbol'])
        risk_amount = self.equity[-1] * risk_value
        pip_value = signal['config']['pip_value']
        size = risk_amount / (2 * signal['atr'] * pip_value)
        size = max(size, 0.01)
        entry_price = signal['price']

        trade = {
            'symbol': signal['symbol'],
            'entry_time': signal['datetime'],
            'entry_price': entry_price,
            'size': size,
            'direction': signal['direction'],
            'tp': None,
            'sl': None,
            'status': 'open',
            'config': signal['config']
        }
        if signal['direction'] == 'buy':
            trade['sl'] = entry_price - 2 * signal['atr']
            trade['tp'] = entry_price + 3 * signal['atr']
        elif signal['direction'] == 'sell':
            trade['sl'] = entry_price + 2 * signal['atr']
            trade['tp'] = entry_price - 3 * signal['atr']

        self.trades.append(trade)
        self.open_trades += 1

        sym = signal['symbol']
        self.trade_counts[sym] = self.trade_counts.get(sym, 0) + 1
        if signal['direction'] == 'buy':
            self.buy_count += 1
        elif signal['direction'] == 'sell':
            self.sell_count += 1
            
    def _partial_close(self, trade: dict, price: float, current_dt: datetime):
        partial_size = trade['size'] * StrategyConfig.PARTIAL_CLOSE_PCT
        if trade['direction'] == 'buy':
            profit = partial_size * (price - trade['entry_price']) * trade['config']['pip_value']
        else:
            profit = partial_size * (trade['entry_price'] - price) * trade['config']['pip_value']
        new_equity = self.equity[-1] + profit
        self.equity.append(new_equity)
        trade.setdefault('realized_profit', 0.0)
        trade['realized_profit'] += profit
        trade['size'] *= (1 - StrategyConfig.PARTIAL_CLOSE_PCT)
        trade['partial_closed'] = True
        trade['partial_close_time'] = current_dt
        if trade['direction'] == 'buy':
            trade['sl'] = trade['entry_price'] + (1 * trade['config']['pip_value'])
        else:
            trade['sl'] = trade['entry_price'] - (1 * trade['config']['pip_value'])

    def _close_trade(self, trade: dict, price: float, reason: str, current_dt: datetime):
        if trade['direction'] == 'buy':
            profit = trade['size'] * (price - trade['entry_price']) * trade['config']['pip_value']
        else:
            profit = trade['size'] * (trade['entry_price'] - price) * trade['config']['pip_value']

        final_leg_pnl = float(profit)
        total_pnl = final_leg_pnl + float(trade.get('realized_partial', 0.0))
        trade['final_leg_pnl'] = final_leg_pnl
        trade['total_pnl'] = total_pnl
        trade['is_loss'] = (total_pnl < 0)

        new_equity = self.equity[-1] + profit
        self.equity.append(new_equity)
        trade['exit_price'] = price
        trade['exit_reason'] = reason
        trade['exit_time'] = current_dt
        trade['status'] = 'closed'
        self.open_trades -= 1

        # (Cooldown set removed)

        self.equity_timeseries[current_dt] = new_equity

        # partial-close outcome metrics
        if trade.get('partial_closed', False):
            if reason == 'TP':
                self.tp_after_partial_count += 1
            elif reason == 'SL':
                if 'partial_close_time' in trade:
                    reverse_time = (current_dt - trade['partial_close_time']).total_seconds()
                    self.sl_after_partial_reverse_times.append(reverse_time)

        # (All bump-threshold updates removed)

    def manage_trades(self, current_dt: datetime, price_data: dict, model, feature_order: list):
        self.update_equity_daily(current_dt)
        for trade in self.trades:
            if trade['status'] != 'open':
                continue
            symbol = trade['symbol']
            symbol_df = price_data[symbol]
            current_row = symbol_df.asof(current_dt)
            if current_row is None or pd.isna(current_row['Close']):
                continue

            current_price = current_row['Close']
            progress = abs(current_price - trade['entry_price']) / (abs(trade['tp'] - trade['entry_price']) + 1e-8)
            if progress >= StrategyConfig.PARTIAL_CLOSE_PCT and not trade.get('partial_closed', False):
                self._partial_close(trade, current_price, current_dt)

            exit_reason = None
            if trade['direction'] == 'buy':
                if current_price >= trade['tp']:
                    exit_reason = 'TP'
                elif current_price <= trade['sl']:
                    exit_reason = 'SL'
            else:
                if current_price <= trade['tp']:
                    exit_reason = 'TP'
                elif current_price >= trade['sl']:
                    exit_reason = 'SL'

            if exit_reason:
                self._close_trade(trade, current_price, exit_reason, current_dt)

# ==================== STRATEGY BACKTEST ====================
class StrategyBacktest:
    def __init__(self):
        self.data_loader = None
        self.engine = None

    def _process_signal(self, df, idx, sym, cfg):
        r = df.iloc[idx]
        feats = {
            'ema_diff':             r['EMA20'] - r['EMA200'],
            'bb_penetration':       (r['Close'] - r['BB_upper']) / r['BB_upper'],
            'bb_lower_penetration': r['bb_lower_penetration'],
            'bb_width':             r['BB_width'],
            'atr_ratio':            r['ATR'] / r['ATR_MA'],
            'adx':                  r['ADX'],
            'rsi14':                r['RSI14'],
            'ema200_slope':         r['EMA200_slope'],
            'vol_ratio':            r['Vol_ratio'],
            'trend_purity':         r['Trend_purity'],
            'ema50_diff':           r['ema50_diff'],
            'EMA50_slope':          r['EMA50_slope'],
            'macd_hist':            r['MACD_hist'],
            'momentum_flag':        r['momentum_flag'],
            'volume_ratio':         r['volume_ratio'],
        }

        # build DataFrame in the exact order your pipeline expects
        order = self.data_loader.feature_order
        X = pd.DataFrame({c: [feats[c]] for c in order}, columns=order)

        probs = self.data_loader.bull_model.predict_proba(X)[0]
        
        if max(probs)<0.6: return  
        
        threshold=0.70
        dire='buy'  if probs[1] > threshold else \
             'sell' if probs[0] > threshold else None
             
        if dire:  
            self.engine.execute_trade({'symbol':sym,'datetime':r.name,  
                'price':r['Close'],'atr':r['ATR'],'config':cfg,  
                'direction':dire,'risk':StrategyConfig.RISK_PER_TRADE})  

    def run(self,start:datetime,end:datetime):  
        price_data={s:d['data'] for s,d in self.data_loader.instruments.items()}  
        for sym,d in self.data_loader.instruments.items():  
            df, cfg = d['data'], d['config']  
            df=df.loc[start:end]  
            for i in range(len(df)):  
                dt=df.index[i]    
                self._process_signal(df,i,sym,cfg)  
                self.engine.manage_trades(dt,price_data,  
                                          self.data_loader.bull_model,  
                                          self.data_loader.feature_order)  
        PerformanceReport(self.engine).generate()  
        
    def run_backtests(self):
        loader = BacktestDataLoader()
        self.data_loader = loader

        # Additional backtest: Entire period from Jan 2014 to Dec 2024
        print("\n=== Backtest: Jan 2024 to Apr 2025 (Entire Period) ===")
        self.start_date = datetime(2024, 1, 1)
        self.end_date = datetime(2025, 4, 11)
        self.engine = ExecutionEngine()
        self.run(self.start_date, self.end_date)

        # Run one backtest per year from 2014 to 2024
        for year in range(2017, 2025):
            print(f"\n=== Backtest: Jan {year} to Dec {year} ===")
            self.start_date = datetime(year, 1, 1)
            self.end_date = datetime(year, 12, 31)
            self.engine = ExecutionEngine()
            self.run(self.start_date, self.end_date)

 

        # Run one additional test for 2025: Jan 1 to Apr 2025
        print("\n=== Backtest: Jan 2025 to apr 2025 ===")
        self.start_date = datetime(2025, 1, 1)
        self.end_date = datetime(2025, 4, 11)
        self.engine = ExecutionEngine()
        self.run(self.start_date, self.end_date)

# ==================== MAIN EXECUTION ====================
if __name__ == "__main__":
    logging.basicConfig(level=logging.INFO)
    backtest = StrategyBacktest()
    backtest.run_backtests()
```
